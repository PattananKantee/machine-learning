### 2. วัตถุประสงค์ของระบบต้นแบบ ###
</br>

### 4. มีค่าที่หายไป (missing values) หรือไม่ อะไรบ้าง แบบใด มากน้อยแค่ไหน แก้ไขได้อย่างไร (ถ้าเป็นรูปภาพ มีรูปที่ไม่ถูกต้องหรือสมบูรณ์หรือไม่ อย่างไร แก้ไขอย่างไร) ###
ข้อมูลชุดนี้มีค่าที่หายไปบางส่วน คือ ข้อมูลรูปภาพและข้อมูลแคปชั่นรวมเป็นจำนวน 3415 ข้อมูล สามารแก้ไขได้โดยการลบแถวของข้อมูลที่หายไปเนื่องจากจำนวนข้อมูลทั้งหมดต่อจำนวนข้อมูลที่หายไปมีค่า 28:1 ซึ่งโดยสัดส่วนนี้ การตัดข้อมูลบางส่วนออกไปนั้นจึงไม่มีผลกระทบ

### 5. แต่ละคุณลักษณะ/feature ต้องมีการทำกระบวนการต่างๆ (operations) หรือไม่ อะไรบ้าง อย่างไร (ถ้าเป็นรูปภาพ ต้องมีลักษณะอย่างไร เช่น ขนาด ความละเอียด รูปแบบไฟล์ ฯลฯ) ###

ข้อมูลรูปภาพชุดนี้สามารถทำกระบวนการ resizing เพื่อแก้ไขขนาดรูปภาพให้มีขนาดเท่ากัน และ pixcel scaling เพื่อปรับความชัดของข้อมูลรูปภาพ

### 6. สำหรับปัญหาและข้อมูลนี้ สามารถป้องกันการเกิด data leakage ได้อย่างไรบ้าง ###
</br>
การเกิด data leakage ของข้อมูลชุดนี้เกิดได้จากรูปภาพบางรูปอาจค้ายคลึงกันหรือเหมือนกัน โดยการป้องกัน data leakage โดยการตัดรูปภาพที่มีความเหมือนหรือคล้ายกันออกไปสามารถทำได้ดังนี้:

- การประมวลผลเบื้องต้น (Preprocessing): เริ่มจากการเตรียมรูปภาพในรูปแบบที่เหมาะสม เช่น การปรับขนาดรูปภาพ (resizing) 

- การคำนวณความคล้ายกัน: ใช้วิธีการคำนวณความคล้ายกันระหว่างเมทริกซ์ของรูปภาพเพื่อหาว่ารูปภาพใดมีความคล้ายกันมากน้อยเพียงใด จากนั้นตั้งเกณฑ์สำหรับการกำหนดว่ารูปภาพใดจะถือว่าคล้ายกันเกินไป และหากพบว่ารูปภาพใดมีความคล้ายกันเกินกว่าที่กำหนด ก็จะทำการตัดรูปภาพนั้นออกจากชุดข้อมูลเพื่อลดความเสี่ยงในการเกิด data leakage

### 7. วิเคราะห์ความสำคัญของคุณลักษณะ ตัวไหนควรและไม่ควรใช้ (ถ้าเป็นรูปภาพ ให้วิเคราะห์ลักษณะของรูปที่ควรและไม่ควรใช้) ###

ลักษณะรูปภาพในชุดข้อมูลที่ควรใช้ คือ รูปภาพที่มี label ซึ่งนั่นก็คือแคปชั่นที่มีความสอดคล้องกับรูปภาพ หรือหมายถึงแคปชั่นที่มีการอธิบายเหตุการณ์แค่เฉพาะที่อยู่ในรูปภาพ ไม่ได้มีการกล่าวถึงสิ่งอื่นๆนอกเหนือจากนั้น

### ประเด็นอื่น ๆ ที่เกี่ยวข้องกับข้อมูลของตัวเอง ###
เนื่องจากมีข้อมูลรูปภาพที่สามารถระบุถึงบุคคลได้ จึงจำเป็นต้องตระหนักถึงจริยธรรมในการใช้ข้อมูลเพื่อนำไปศึกษา, จากชุดข้อมูลเราไม่สามารถทราบได้ว่ารูปภาพนั้นได้รับการยินยอมให้นำมาใช้เพื่อการศึกษาหรือไม่
