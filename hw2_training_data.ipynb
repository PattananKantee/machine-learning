### 3. ข้อมูลมีการ/น่าจะมีการสุ่มตัวอย่างมา (sampling) แบบใด เพราะเหตุใด มีปัญหาหรือไม่ อย่างไร แก้ไขได้อย่างไร ###
</br>
ชุดข้อมูล Instagram Images with Captions Datasets มีการสุ่มตัวอย่างมาแบบ Snowball sampling เพราะ ข้อมูลทั้งสามชุดเป็นการเก็บข้อมูลจากแพลตฟอร์มอินสตาแกรม ซึ่งเป็นโซเชียลเน็ตเวิร์กจึงสามารถเก็บข้อมูลได้จากความเชื่อมโยงของผู้ใช้งานได้อย่างง่าย โดยการสุ่มข้อมูลในรูปแบบนี้อาจมีปัญหาดังนี้
</br>
</br>
1. การสุ่มตัวอย่างแบบ Snowball sampling ไม่มีการควบคุมที่ชัดเจนในการเลือกตัวอย่าง 
</br>
2. มีการเก็บข้อมูลไม่ทั่วถึงเพราะมีภาวะอิสระต่ำ 
</br>
</br>
มีวิธีแก้ไขได้โดยการควบคุมการสุ่มตัวอย่าง เช่น การเลือกผู้สมัครเริ่มต้นจากพื้นที่ที่กลุ่มเป้าหมายต่อกลุ่มประชากรที่ต้องการศึกษา และเพิ่มความหลากหลายในการเลือกตัวอย่างเพื่อลดความเอนเอียงของข้อมูล

### 4. ข้อมูลมีการ/น่าจะมีการกำหนดเฉลย (labeling) ด้วยวิธีใด เพราะเหตุใด มีปัญหาหรือไม่ อย่างไร แก้ไขได้อย่างไร ###
</br>
</br>
ชุดข้อมูลนี้มีการกำหนดเฉลยด้วยวิธี Hand Labels เพราะ ข้อมูลที่ได้นั้นมาจากแพลตฟอร์มอินตาแกรม ถึงแม้จะเป็นการเฉลยด้วยวิธี Hand Labels ที่โดยส่วนใหญ่แล้วจะมีค่าใช้จ่ายมากและใช้เวลานานในการกำหนดเฉลย แต่เนื่องจากแพลตฟอร์มอินตาแกรมมีผู้ใช้งานจำนวนมากทำให้มีข้อมูลเพียงพอ ซึ่งผู้ใช้งานได้กำหนดเฉลยไว้พร้อมข้อมูลรูปภาพ และได้เลือกที่จะเปิดเผยรูปภาพและแคปชันไว้แล้วจึงไม่มีปัญหาในด้านความเป็นส่วนตัว 

### 5. ข้อมูลขาดความสมดุล (imbalance) หรือไม่ อย่างไร มากน้อยแค่ไหน มีปัญหาหรือไม่ อย่างไร แก้ไขได้อย่างไร ###
</br>
ข้อมูลขาดความสมดุลมาก โดยที่ข้อมูลส่วนใหญ่เป็นรูปภาพคน และไม่ค่อยมีรูปประเภทอื่นๆ จึงเลือกที่จะแก้ไขโดยการแยกประเภทรูปภาพออกเป็นรูปภาพที่มีหน้าคนและรูปภาพที่ไม่มีหน้าคน และโมเดลของเราจะเลือกใช้เฉพาะรูปภาพที่มีหน้าคนและตัดรูปอื่นๆออกไป ซึ่งเราจะทำการนำเข้าโมเดล https://huggingface.co/microsoft/resnet-50 นี้ที่ช่วยในการแบ่งประเภทรูปภาพเพื่อแยกข้อมูลรูปภาพของเราฃ

### 6. จำเป็นหรือสามารถสร้างข้อมูลเพิ่ม (augmentation) ได้หรือไม่ อย่างไร ###
</br>
จำเป็นต้องสร้างข้อมูลเพิ่มโดยวิธีการดังนี้
</br>
1. Simple Label-Preserving Transformations: มีการเปลี่ยนแปลงข้อมูลรูปภาพโดยการหมุนหรือพลิกรูปภาพ
</br>
2. Perturbation: เพิ่มความไม่เสถียรของภาพ (noise) 



### ประเด็นอื่น ๆ ที่เกี่ยวข้องกับข้อมูลของตัวเอง ###
เนื่องจากมีข้อมูลรูปภาพที่สามารถระบุถึงบุคคลได้ จึงจำเป็นต้องตระหนักถึงจริยธรรมในการใช้ข้อมูลเพื่อนำไปศึกษา, จากชุดข้อมูลเราไม่สามารถทราบได้ว่ารูปภาพนั้นได้รับการยินยอมให้นำมาใช้เพื่อการศึกษาหรือไม่
