### ภาพจาก kaggle ใช้ได้ เพราะมี path file ในเครื่อง

```python

image1 = cv2.imread("img/insta229.jpg")
image2 = cv2.imread("img/insta231.jpg")
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]), interpolation = cv2.INTER_AREA)
print(image1.shape, image2.shape)
# Convert images to grayscale
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
# Calculate SSIM
ssim_score = metrics.structural_similarity(image1_gray, image2_gray, full=True)
print(f"SSIM Score: ", ssim_score[0])

#if ssim = 1 : เหมือน
#if ssim < 1 : ไม่เหมือน
```

### ภาพจาก hugging face ใช้ไม่ได้ ไม่มีไฟล์

```python
from datasets import load_dataset, Image

dataset = load_dataset("mrSoul7766/instagram_post_captions", split="train")
dataset[0]["image"]
```

```python

image1 = cv2.imread(dataset[0]["image"])
image2 = cv2.imread(dataset[1]["image"])
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]), interpolation = cv2.INTER_AREA)
print(image1.shape, image2.shape)
# Convert images to grayscale
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
# Calculate SSIM
ssim_score = metrics.structural_similarity(image1_gray, image2_gray, full=True)
print(f"SSIM Score: ", ssim_score[0])

#if ssim = 1 : เหมือน
#if ssim < 1 : ไม่เหมือน
```
